# **1️⃣ Which Team Has Won the Most IPL Matches?**


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
matches = pd.read_csv(r"C:\Users\SANTOSH\Downloads\matches.csv")

# Count number of matches won by each team
team_wins = matches['winner'].value_counts()

# Display the result
print("Number of matches won by each team:")
print(team_wins)

# Plot the result
plt.figure(figsize=(12,6))
sns.barplot(x=team_wins.index, y=team_wins.values, color='skyblue')  # Use color instead of palette
plt.xticks(rotation=90)
plt.title("Most IPL Matches Won by Teams", fontsize=15)
plt.xlabel("Team", fontsize=12)
plt.ylabel("Number of Matches Won", fontsize=12)
plt.show()
```

# **Which Season Had the Most Matches Played?**


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
matches = pd.read_csv(r"C:\Users\SANTOSH\Downloads\matches.csv")

# Count matches per season
season_matches = matches['season'].value_counts().sort_index()

# Convert to DataFrame for seaborn
season_df = season_matches.reset_index()
season_df.columns = ['season', 'matches']

# Plot with different colors
plt.figure(figsize=(12,6))
sns.barplot(data=season_df, x='season', y='matches', hue='season', palette="viridis", legend=False)

plt.title("Number of Matches Played in Each IPL Season", fontsize=16, fontweight='bold')
plt.xlabel("Season", fontsize=12)
plt.ylabel("Number of Matches", fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

# How Often Did the Toss Winner Win the Match?


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv(r"C:\Users\SANTOSH\Downloads\matches.csv")

# Create a new column: Did the toss winner also win the match?
df['toss_win_match_win'] = df['toss_winner'] == df['winner']

# Count the results
result_counts = df['toss_win_match_win'].value_counts()

# Prepare data for visualization
labels = ['Toss Winner Also Won Match', 'Toss Winner Lost Match']
sizes = [result_counts[True], result_counts[False]]
colors = ['#4CAF50', '#F44336']

# Plot a pie chart
plt.figure(figsize=(8,8))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140, textprops={'fontsize': 12})
plt.title('How Often Did the Toss Winner Win the Match?', fontsize=16, fontweight='bold')
plt.show()
```

# **3️⃣ Which City Hosted the Most IPL Matches?**


```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Dataset
# Replace 'matches.csv' with the path to your IPL matches dataset
df = pd.read_csv("C:\\Users\\SANTOSH\\Downloads\\matches.csv")



# Group by city and count the number of matches
city_matches = df['city'].value_counts()

# Visualization
plt.figure(figsize=(12,6))
sns.barplot(x=city_matches.index, y=city_matches.values, color="skyblue")

plt.title('Number of IPL Matches Hosted in Each City', fontsize=16, fontweight='bold')
plt.xlabel('City', fontsize=14)
plt.ylabel('Number of Matches', fontsize=14)
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()

```
